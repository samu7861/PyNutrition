```python
#entradas
name = input("ingrese nombre del paciente")
a = input("Ingrese el género (m)(f)")
genero = a.lower()
peso = int(input("ingrese peso"))
altura = int(input("ingrese altura en cm"))
edad = int(input("ingrese edad"))
FA = input("Factor de actividad (S=sedentario)(L=ligero)(M=moderado)(I=intensa)")
fa=FA.lower()
OBJETIVO = input("Qué objetivo tiene? (1=perdia de grasa)(2=aumento de masa muscular)(3=mantenimiento de peso)(4=perdida de peso)(5=aumento de peso)")
RANGO = float(input("Ingrese el rango de proteína"))
ReqGrasa = int(input("Qué porcentaje de requerimiento de grasa"))

print("-------------------")

#Magnitudes derivadas
Talla=altura/100
print("Talla = ",Talla, "m")
IMC=round(peso/Talla**2)
print("IMC = ", IMC)
ReqProt=RANGO*peso
print("El requerimiento proteico es = ", ReqProt,"gr")
CalProt=ReqProt*4
print("Calorías de proteina= ",CalProt )


#funciones
def TipoAdulto(edad):
  if edad>59:
    TA="Adulto Mayor"
  else:
    TA="Adulto"
  return TA

Resultado_TipoAdulto = TipoAdulto(edad)
print(Resultado_TipoAdulto)


def CLASIFICACION(IMC, Resultado_TipoAdulto):
  if Resultado_TipoAdulto == "Adulto Mayor":
    if IMC <= 16:
      clfn = "Desnutrición severa"
    if IMC > 16 and IMC <= 16.9:
      clfn = "Desnutrición moderada"
    if IMC > 16.9 and IMC <=18.4:
      clfn = "Desnutrición leve"
    if IMC > 18.4 and IMC <=22:
      clfn = "Peso insuficiente"
    if IMC > 22 and IMC <=27:
      clfn = "Eutrófico"
    if IMC > 27 and IMC <=29.9:
      clfn = "Sobrepeso"
    if IMC > 29.9 and IMC <=34.9:
      clfn = "Obesidad 1"
    if IMC > 34.9 and IMC <=39.9:
      clfn = "Obesidad 2"
    if IMC > 39.9 and IMC <=49.9:
      clfn = "Obesidad 3"
    if IMC >=50:
      clfn = "Obesidad mórbida"
  if Resultado_TipoAdulto == "Adulto":
    if IMC <= 16:
      clfn = "Desnutrición severa"
    if IMC > 16 and IMC <= 16.9:
      clfn = "Desnutrición moderada"
    if IMC > 16.9 and IMC <=18.4:
      clfn = "Desnutrición leve"
    if IMC > 18.4 and IMC <=24.9:
      clfn = "Eutrófico"
    if IMC > 24.9 and IMC <=29.9:
      clfn = "Sobrepeso"
    if IMC > 29.9 and IMC <=34.9:
      clfn = "Obesidad 1"
    if IMC > 34.9 and IMC <=39.9:
      clfn = "Obesidad 2"
    if IMC > 39.9 and IMC <=44.9:
      clfn = "Obesidad 3"
    if IMC >= 45:
      clfn = "Obesidad mórbida"
  return clfn

Resultado_CLASIFICACION = CLASIFICACION(IMC, Resultado_TipoAdulto)
print(Resultado_CLASIFICACION)

    
    


def GEBASAL(peso, altura, edad, genero):
  if genero == "m":
    Gbasal = (10*peso)+(6.25*altura)-(5*edad)+5
  else:
    Gbasal = (10*peso)+(6.25*altura)-(5*edad)-161
  return Gbasal
  
resultado_GEBASAL = GEBASAL(peso, altura, edad, genero)
print("El resultado de GEBASAL es:", resultado_GEBASAL)


def GETOTAL(resultado_GEBASAL, FA, OBJETIVO):
  if OBJETIVO=="5":
    if fa=="i":
      Gtotal = resultado_GEBASAL*1.725+250
    if fa=="m":
      Gtotal = resultado_GEBASAL*1.55+250
    if fa=="l":
      Gtotal = resultado_GEBASAL*1.375+250
    else:
      Gtotal = resultado_GEBASAL*1.2+250
  if OBJETIVO=="2":
    if fa=="i":
      Gtotal = resultado_GEBASAL*1.725+200
    if fa=="m":
      Gtotal = resultado_GEBASAL*1.55+200
    if fa=="l":
      Gtotal = resultado_GEBASAL*1.375+200
    else:
      Gtotal = resultado_GEBASAL*1.2+200
  else:
    if fa=="i":
      Gtotal = resultado_GEBASAL*1.725
    if fa=="m":
      Gtotal = resultado_GEBASAL*1.55
    if fa=="l":
      Gtotal = resultado_GEBASAL*1.375
    else:
      Gtotal = resultado_GEBASAL*1.2
  return Gtotal


resultado_GETOTAL = GETOTAL(resultado_GEBASAL, FA, OBJETIVO)
print("El resultado de GETOTAL es:", resultado_GETOTAL)


def DFTcal(resultado_GETOTAL, IMC, Resultado_TipoAdulto, CLASIFICACION):
  if IMC > 24.9:
    dftcal=resultado_GETOTAL-250
  else:
    dftcal=resultado_GETOTAL
  return dftcal


resultado_DFTcal = GETOTAL(resultado_GEBASAL, FA, OBJETIVO)
print("Deficit calórico: ", resultado_DFTcal)


CalGrasa=round(ReqGrasa*resultado_DFTcal/100)
print("Calorías de grasa = ", CalGrasa)

GrGrasa=round(CalGrasa/9)
print("Gramos de grasa = ", GrGrasa, "g")

CalCarb=round(resultado_DFTcal-CalProt-CalGrasa)
print("Calorías de Carbohidratos = ", CalCarb)

GrCarb=round(CalCarb/4)
print("gramos de Carbohidratos = ", GrCarb,"g")

ReqAgua=peso*35
vasos=round(ReqAgua/200)
print("Requerimiento de agua = ", ReqAgua)
print("Vasos de agua = ", vasos)
```
